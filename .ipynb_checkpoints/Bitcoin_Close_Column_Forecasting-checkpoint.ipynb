```python
def rolling_forecast_naive(df, train_len, test_len, window):
    total_len = train_len + test_len
    naive_pred = []
    for j in np.arange(train_len, total_len, window):
        data = df.iloc[:j]
        pred_value = data.values.ravel()[-1]
        naive_pred.extend(pred_value for _ in range(window))
    naive_pred = naive_pred[:test_len]
    return naive_pred
```

```python
def rolling_forecast_mean(df, train_len, test_len, window):
    total_len = train_len + test_len
    mean_pred = []
    for j in np.arange(train_len, total_len, window):
        data = df.iloc[:j]
        pred_value = np.mean(data.values)
        mean_pred.extend(pred_value for _ in range(window))
    mean_pred = mean_pred[:test_len]
    return mean_pred
```

```python
def rolling_forecast_AR(df, train_len, test_len, window):
    total_len = train_len + test_len
    ar_pred = []

    ps = range(0, 3)  

    # Function to find the best AR parameter for the current data window
    def find_best_ar_param(data, ps):
        best_aic = np.inf
        best_p = 1
        for p in ps:
            try:
                model = ARIMA(data, order=(p, 0, 0))
                result = model.fit()
                if result.aic < best_aic:
                    best_aic = result.aic
                    best_p = p
            except:
                continue
        return best_p

    for j in np.arange(train_len, total_len, window):
        data = df.iloc[:j]

        # Find the best AR parameter for the current window of data
        best_p = find_best_ar_param(data, ps)

        # Fit model with the best parameters found
        model = ARIMA(data, order=(best_p, 0, 0))
        result = model.fit()
        pred_value = result.get_forecast(steps=window)
        pred_mean = pred_value.predicted_mean
        # Handling the prediction length due to window size
        if len(pred_mean) > window:
            pred_mean = pred_mean[:window]
        ar_pred.extend(pred_mean)

    ar_pred = ar_pred[:test_len]  # Ensure we only return the required number of predictions
    return ar_pred

```

```python
def rolling_forecast_MA(df, train_len, test_len, window):
    total_len = train_len + test_len
    ma_pred = []

    qs = range(0, 3)  
    
    # Function to find the best MA parameter for the current data window
    def find_best_ma_param(data, qs):
        best_aic = np.inf
        best_q = 0
        for q in qs:
            try:
                model = SARIMAX(data, order=(0, 0, q))
                result = model.fit(disp=False)
                if result.aic < best_aic:
                    best_aic = result.aic
                    best_q = q
            except:
                continue
        return (0, 0, best_q)  # Since p and d are fixed at 0

    for j in np.arange(train_len, total_len, window):
        data = df.iloc[:j]

        # Find the best MA parameter for the current window of data
        best_order = find_best_ma_param(data, qs)

        # Fit model with the best parameters found
        model = SARIMAX(data, order=best_order)
        result = model.fit(disp=False)
        pred_value = result.get_prediction(start=j, end=j + window - 1)
        ma_pred.extend(pred_value.predicted_mean)

    ma_pred = ma_pred[:test_len]  # Ensure we only return the required number of predictions
    return ma_pred
```

```python
pred_df = test.copy()
train_len = len(train)
test_len = len(test)
window = 1
pred_naive_day = rolling_forecast_naive(data_diff, train_len, test_len, window)
pred_mean_day = rolling_forecast_mean(data_diff, train_len, test_len, window)
pred_AR_day = rolling_forecast_AR(data_diff, train_len, test_len, window)
pred_MA_day = rolling_forecast_MA(data_diff, train_len, test_len, window)
pred_df['pred_naive_day'] = pred_naive_day
pred_df['pred_mean_day'] = pred_mean_day
pred_df['pred_AR_day'] = pred_AR_day
pred_df['pred_MA_day'] = pred_MA_day
```

```python
from prophet import Prophet

# Load and prepare the data
data = df[['Date', 'Close']]  # Ensure df is defined or loaded before this script
data['Date'] = pd.to_datetime(data['Date'])
prophet_data = data[['Date', 'Close']].rename(columns={'Date': 'ds', 'Close': 'y'})

# Drop any rows with NaN values to ensure clean data for modeling
prophet_data.dropna(inplace=True)

# Initialize and fit the Prophet model
model = Prophet(daily_seasonality=False)  # Adjust seasonality as needed
model.fit(prophet_data)

# Create future dataframe for forecasting
future = model.make_future_dataframe(periods=30)  # Extend future dates beyond the last date in dataset

# Make predictions
forecast = model.predict(future)

# Filter forecast to dates that overlap with the historical data for visualization
forecast_filtered = forecast[['ds', 'yhat']].set_index('ds')
prophet_preds_test = forecast_filtered.join(prophet_data.set_index('ds'), how='inner')['yhat']

# Plot the forecast
fig = model.plot(forecast)
plt.title('Forecast of Close Prices')
plt.show()

# Plot components of the forecast (trends, yearly seasonality, etc.)
fig2 = model.plot_components(forecast)
plt.show()

# Plotting actual vs predicted values for visual comparison
plt.figure(figsize=(10, 5))
plt.plot(prophet_data['ds'], prophet_data['y'], label='Actual Close Prices')
plt.plot(prophet_preds_test.index, prophet_preds_test, label='Predicted Close Prices', linestyle='--')
plt.title('Actual vs Predicted Close Prices')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Assuming 'bitcoin_data' is your DataFrame with Bitcoin time series data

# Step 1: Data Preprocessing
# Drop unnecessary columns
bitcoin_data = df[['Date', 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']]
# Convert 'Date' column to datetime
bitcoin_data['Date'] = pd.to_datetime(bitcoin_data['Date'])

# Step 2: Feature Selection
# Assuming 'Open', 'Low', 'Close', 'AdjClose', and 'Volume' are used as features

# Step 3: Split the Data
X = bitcoin_data[['Open', 'Low', 'High', 'Adj Close', 'Volume']]
y = bitcoin_data['Close']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Step 4: Choose Regression Models
# SVR
svr_model = SVR(kernel='linear')

# Random Forest Regression
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)

# XGBoost Regression
xgb_model = XGBRegressor()

# Step 5: Train the Models
# Scale the features for SVR
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train SVR
svr_model.fit(X_train_scaled, y_train)

# Train Random Forest Regression
rf_model.fit(X_train, y_train)

# Train XGBoost Regression
xgb_model.fit(X_train, y_train)

# Step 6: Evaluate the Models
# Predictions
svr_preds_train = svr_model.predict(X_train_scaled)
svr_preds_test = svr_model.predict(X_test_scaled)

rf_preds_train = rf_model.predict(X_train)
rf_preds_test = rf_model.predict(X_test)

xgb_preds_train = xgb_model.predict(X_train)
xgb_preds_test = xgb_model.predict(X_test)

# Calculate evaluation metrics for SVR
svr_mae_train = mean_absolute_error(y_train, svr_preds_train)
svr_mae_test = mean_absolute_error(y_test, svr_preds_test)

svr_mse_train = mean_squared_error(y_train, svr_preds_train)
svr_mse_test = mean_squared_error(y_test, svr_preds_test)

svr_r2_train = r2_score(y_train, svr_preds_train)
svr_r2_test = r2_score(y_test, svr_preds_test)

# Calculate evaluation metrics for Random Forest Regression
rf_mae_train = mean_absolute_error(y_train, rf_preds_train)
rf_mae_test = mean_absolute_error(y_test, rf_preds_test)

rf_mse_train = mean_squared_error(y_train, rf_preds_train)
rf_mse_test = mean_squared_error(y_test, rf_preds_test)

rf_r2_train = r2_score(y_train, rf_preds_train)
rf_r2_test = r2_score(y_test, rf_preds_test)


# Calculate evaluation metrics for XGBoost Regression
xgb_mae_train = mean_absolute_error(y_train, xgb_preds_train)
xgb_mae_test = mean_absolute_error(y_test, xgb_preds_test)

xgb_mse_train = mean_squared_error(y_train, xgb_preds_train)
xgb_mse_test = mean_squared_error(y_test, xgb_preds_test)

xgb_r2_train = r2_score(y_train, xgb_preds_train)
xgb_r2_test = r2_score(y_test, xgb_preds_test)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Function to create visualizations
def plot_results(actual, predicted, title):
    plt.figure(figsize=(10, 6))
    plt.scatter(actual, predicted, alpha=0.5)
    plt.plot([actual.min(), actual.max()], [actual.min(), actual.max()], 'k--', lw=2)
    plt.xlabel('Actual')
    plt.ylabel('Predicted')
    plt.title(title)
    plt.grid(True)
    plt.show()

# Plotting actual vs. predicted values
plot_results(y_train, svr_preds_train, "SVR: Actual vs. Predicted (Training Set)")
plot_results(y_train, rf_preds_train, "Random Forest: Actual vs. Predicted (Training Set)")
plot_results(y_train, xgb_preds_train, "XGBoost: Actual vs. Predicted (Training Set)")


plot_results(y_test, svr_preds_test, "SVR: Actual vs. Predicted (Testing Set)")
plot_results(y_test, rf_preds_test, "Random Forest: Actual vs. Predicted (Testing Set)")
plot_results(y_test, xgb_preds_test, "XGBoost: Actual vs. Predicted (Testing Set)")


# Function to plot residuals
def plot_residuals(actual, predicted, title):
    residuals = actual - predicted
    plt.figure(figsize=(10, 6))
    sns.histplot(residuals, kde=True)
    plt.xlabel('Residuals')
    plt.ylabel('Frequency')
    plt.title(title)
    plt.grid(True)
    plt.show()

# Plotting residuals
plot_residuals(y_train, svr_preds_train, "SVR: Residuals Distribution (Training Set)")
plot_residuals(y_train, rf_preds_train, "Random Forest: Residuals Distribution (Training Set)")
plot_residuals(y_train, xgb_preds_train, "XGBoost: Residuals Distribution (Training Set)")


plot_residuals(y_test, svr_preds_test, "SVR: Residuals Distribution (Testing Set)")
plot_residuals(y_test, rf_preds_test, "Random Forest: Residuals Distribution (Testing Set)")
plot_residuals(y_test, xgb_preds_test, "XGBoost: Residuals Distribution (Testing Set)")
```
